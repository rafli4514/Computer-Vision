```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import os

file_path = "/home/rafli/semester_6/Computer_Vision/Computer-Vision/Images/images.jpg"

image = os.path.basename(file_path)
image = Image.open(file_path).convert('RGB')
image_array = np.array(image, dtype=np.int32)
```

```python
print(image_array)
```

```python
height = image_array.shape[0]
width = image_array.shape[1]
```

```python
print(height)
print(width)
```

```python
t = np.zeros((height, width, 3), dtype=np.uint8)
L = 256

for i in range(height):
    for j in range(width):
        for k in range(3):
            I = image_array[i][j][k]
            t[i][j][k] = L - 1 - I
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image_array)
axes[0].set_title("Original")
axes[0].axis('off')
axes[1].imshow(t)
axes[1].set_title("Inverted Image")
axes[1].axis('off')
plt.tight_layout()
plt.show()
```

```python
height, width, channels = image_array.shape
```

```python
T = np.zeros((height, width, channels), dtype=np.int32)

for i in range(height):
    for j in range(width):
        for k in range(3):
            T[i][j][k] = image_array[i][j][k] + 50
            T = np.clip(T, 0, 255)

print(T)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image)
axes[0].set_title("Original")
axes[0].axis('off')
axes[1].imshow(T)
axes[1].set_title("Brightness filter")
axes[1].axis('off')
```
