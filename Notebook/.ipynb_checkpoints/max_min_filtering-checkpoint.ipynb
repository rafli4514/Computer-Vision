```python
import matplotlib.pyplot as plt
import scipy
from scipy import ndimage
import scipy.ndimage
from PIL import Image
```

```python
file_path = 'Images/Screenshot 2025-03-11 091107.png'

image = Image.open(file_path).convert('L')
```

```python
filter_image_max = ndimage.maximum_filter(
    image, 
    size=3, 
    footprint=None,
    output=None, 
    mode='reflect', 
    cval=0.0, 
    origin=0, 
    axes=None
)

filter_image_min = ndimage.minimum_filter(
    image, 
    size=3, 
    footprint=None,
    output=None, 
    mode='reflect', 
    cval=0.0, 
    origin=0, 
    axes=None
)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original")
axes[0].axis('off')
axes[1].imshow(filter_image_max, cmap='gray')
axes[1].set_title("Max filter")
axes[1].axis('off')
axes[2].imshow(filter_image_min, cmap='gray')
axes[2].set_title("Min filter")
axes[2].axis('off')
```
